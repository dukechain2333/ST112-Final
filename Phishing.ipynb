```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
```

```python
data = pd.read_csv('Phishing.csv')
data
```

```python
data.columns
```

```python
data = data.dropna(axis = 0)
y = data.pop('URL_Type_obf_Type')
X = data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
X_train
```

```python
clf = RandomForestClassifier(n_estimators=80, max_depth=85)
clf.fit(X_train, y_train)
```

```python
k = 5
n_estimators_range = np.arange(80,90,1)
max_depth_range = np.arange(80,90,1)
params = {
    'max_depth': max_depth_range,
    'n_estimators': n_estimators_range
}

grid_model = GridSearchCV(estimator=clf,
                          param_grid=params,
                          cv=k,
                          return_train_score=True,
                          scoring='accuracy'                        )

grid_model_result = grid_model.fit(X_train, y_train)
```

```python
cv_results = pd.DataFrame(grid_model.cv_results_)
cv_results.head(10)
```
