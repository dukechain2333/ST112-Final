```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
from sklearn.ensemble import RandomForestClassifier
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv('Defacement_Infogaincsv')
data.head(10)
```

```python
data.columns
```

```python
y = (data.pop('class'))


X = data
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=1234)
X_train
```

```python
clf = RandomForestClassifier(n_estimators=85, max_depth=85)
clf.fit(X_train, y_train)
```

```python
label_predict = clf.predict(X_test)
label_predict
```

```python
accuracy_score(y_test, label_predict)
```

```python
pca = PCA(n_components=5).fit(X_train)
pca_data_train = pca.transform(X_train)
pca_data_test = pca.transform(X_test)
pca_data_train
```

```python
clf = RandomForestClassifier(max_depth=10)
clf.fit(pca_data_train, y_train)
label_predict = clf.predict(pca_data_test)
accuracy_score(y_test, label_predict)
```
